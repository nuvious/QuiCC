```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

C_values = [0.125 * 2**i for i in range(5)]  # Starting with 0.125 and doubling up to 2
M_values = [8 * 2**i for i in range(10)]  # Starting with 8 and doubling up to 1048576
M_values_log2 = np.log2(M_values)
M_log2, C = np.meshgrid(M_values_log2, C_values)
Z = (8 * (2**M_log2)) * C / (512 + 16 + (2**M_log2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(C, M_log2, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('Connection Rate (Connections/Sec)')
ax.set_ylabel('log2(Message Size (Bytes))')
ax.set_zlabel('Bandwidth (Bytes/Sec)')
ax.set_title('QUIC Connection ID Based Covert Channel Bandwidth')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
